```python
import cv2
import torch
import matplotlib.pyplot as plt
```

### Contenido

1. Cargar una imagen de prueba usando OpenCV.

2. Convertirla de BGR (el formato de OpenCV) a RGB (el formato estándar).

3. Convertirla a un tensor de PyTorch.

4. Usar Matplotlib para mostrar la imagen original.

```python
image_path = '../data/test_image.jpg'
try:
    image_cv = cv2.imread(image_path)
    if image_cv is None:
        raise FileNotFoundError(f"No se pudo cargar la imagen en: {image_path}")

    # 2. Convertir de BGR a RGB
    image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

    # 3. Convertir la imagen de NumPy array a un tensor de PyTorch
    # PyTorch espera las dimensiones en (Canales, Alto, Ancho)
    tensor_pytorch = torch.from_numpy(image_rgb).permute(2, 0, 1)

    print("¡Setup verificado con éxito!")
    print(f"Dimensiones de la imagen (OpenCV - H, W, C): {image_rgb.shape}")
    print(f"Dimensiones del tensor (PyTorch - C, H, W): {tensor_pytorch.shape}")
    print(f"Tipo de dato del tensor: {tensor_pytorch.dtype}")

    # 4. Mostrar la imagen usando Matplotlib
    plt.imshow(image_rgb)
    plt.title("Imagen de Prueba Cargada")
    plt.axis('off') # Ocultar los ejes
    plt.show()

except FileNotFoundError as e:
    print(e)
    print("Por favor, asegúrate de que la imagen exista y la ruta sea correcta.")
```
